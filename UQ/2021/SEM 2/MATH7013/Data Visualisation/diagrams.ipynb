```python
import matplotlib.pyplot as plt
import matplotlib.dates as dates
import pandas as pd

# Set font size when plotting
plt.rcParams.update({'font.size': 25})
```

### Load in the preprocessed datasets

```python
# S&P 500 in 2017
df_1 = pd.read_csv("../Data Preprocessing/SPXUSD_2017.csv", index_col="date")
# Convert date column to datetime type
df_1.index = pd.to_datetime(df_1.index)
df_1.head()
```

```python
# NIKKEI 225 in 2017
df_2 = pd.read_csv("../Data Preprocessing/JPXJPY_2017.csv", index_col="date")
# Convert date column to datetime type
df_2.index = pd.to_datetime(df_2.index)
df_2.head()
```

```python
# DAX 30 in 2017
df_3 = pd.read_csv("../Data Preprocessing/GRXEUR_2017.csv", index_col="date")
# Convert date column to datetime type
df_3.index = pd.to_datetime(df_3.index)
df_3.head()
```

### Plot raw historical prices for each dataset

```python
fig, axes = plt.subplots(3, 1, figsize=(20, 14))

axes[0].plot(df_1.index, df_1["p_t"], color="blue")
axes[0].set_xlabel("Date")
axes[0].set_ylabel("Price ($ USD)")
axes[0].set_title("S&P 500: 1-Minute Price for 2017")

axes[1].plot(df_2.index, df_2["p_t"], color="red")
axes[1].set_xlabel("Date")
axes[1].set_ylabel("Price (¥ YEN)")
axes[1].set_title("NIKKEI 225: 1-Minute Price for 2017")

axes[2].plot(df_3.index, df_3["p_t"], color="green")
axes[2].set_xlabel("Date")
axes[2].set_ylabel("Price (€ EUR)")
axes[2].set_title("DAX 30: 1-Minute Price for 2017")

fig.tight_layout()
# fig.savefig("p_t.png", facecolor="w")
```

### Plot raw 1-minute price changes for each dataset

```python
fig, axes = plt.subplots(3, 1, figsize=(20, 14))

axes[0].plot(df_1.index, df_1["z_t"], color="blue")
axes[0].set_xlabel("Date")
axes[0].set_ylabel("$z_t$ ($ USD)")
axes[0].set_title("S&P 500: 1-Minute Price Change for 2017")

axes[1].plot(df_2.index, df_2["z_t"], color="red")
axes[1].set_xlabel("Date")
axes[1].set_ylabel("$z_t$ (¥ YEN)")
axes[1].set_title("NIKKEI 225: 1-Minute Price Change for 2017")

axes[2].plot(df_3.index, df_3["z_t"], color="green")
axes[2].set_xlabel("Date")
axes[2].set_ylabel("$z_t$ (€ EUR)")
axes[2].set_title("DAX 30: 1-Minute Price Change for 2017")

fig.tight_layout()
# fig.savefig("z_t.png", facecolor="w")
```

### Plot raw 10-day momentum change for each dataset

```python
fig, axes = plt.subplots(3, 1, figsize=(20, 14))

axes[0].plot(df_1.index, df_1["m_(t-14400)"], color="blue")
axes[0].set_xlabel("Date")
axes[0].set_ylabel("$m_(t-14400)$ ($ USD)")
axes[0].set_title("S&P 500: 10-Day Momentum Change for 2017")

axes[1].plot(df_2.index, df_2["m_(t-14400)"], color="red")
axes[1].set_xlabel("Date")
axes[1].set_ylabel("$m_(t-14400)$ (¥ YEN)")
axes[1].set_title("NIKKEI 225: 10-Day Momentum Change for 2017")

axes[2].plot(df_3.index, df_3["m_(t-14400)"], color="green")
axes[2].set_xlabel("Date")
axes[2].set_ylabel("$m_(t-14400)$ (€ EUR)")
axes[2].set_title("DAX 30: 10-Day Momentum Change for 2017")

fig.tight_layout()
# fig.savefig("m_(t-14400).png", facecolor="w")
```
